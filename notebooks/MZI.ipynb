# MZI using GDSFactory
This notebook explores building and simulating a Machâ€“Zehnder interferometer (MZI) using [GDS Factory](https://gdsfactory.github.io/gdsfactory/index.html).



```python
import gdsfactory as gf
gf.gpdk.PDK.activate()

# Check Python executable ("where this code is running")
import sys
sys.executable
```

Create a GDS Factory component.

```python
MZI = gf.Component()
```

Create the MZI geometry.

```python
mzi = gf.components.mzi()
MZI << mzi # same as MZI.add_ref(mzi)
MZI.plot()
```
